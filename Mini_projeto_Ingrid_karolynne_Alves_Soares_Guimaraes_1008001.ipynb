```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def jogada(): #função para somar os valores de um lançamento com 2 dados
  dado1, dado2 = np.random.randint(1, 7), np.random.randint(1, 7)
  return dado1 + dado2
```

```python
def simulacoes_jogadas(n: int):
    somas = np.array([jogada() for i in range(n)])
    return somas
```

```python
def analise(jogos): #fução recebe uma np.array com as jogadas imprime e retorna a frequencia de cada resultado
    cont, bordas = np.histogram(jogos, bins=np.arange(2, 14))
    analise = [print(f'Lançamentos com resultado {i+2} ocorreram {quant} vezes!') for i, quant in enumerate(cont)]
    return cont
```

```python
#simulação de 1000 jogadas
quantidade_jogadas = 1000
jogos = simulacoes_jogadas(quantidade_jogadas)

# Mínimo, Máximo e Média dos resultados
media, minimo, maximo = jogos.mean(), jogos.min(), jogos.max()
print(f'A média dos valores dos jogos foi de: {media}\nO valor mínimo dos jogos foi de: {minimo}\nO valor maximo dos jogos foi de: {maximo}\n')

# Número de vezes que cada lançamento ocorreu
resultado = analise(jogos)
```

```python
# Criando um gráfico de histograma das frequencias dos lançamentos dos dados
plt.bar(np.arange(2, 13), resultado)
plt.xlabel('Resultado dos dados')
plt.ylabel('Frequência dos resultados')
plt.title('Histograma das jogadas')
plt.show()
```

# Teste de Hipótese

**Hipótese: Todos os lançamentos são igualmente prováveis.**

Essa afirmação não é consistente com os resultados da simulação realizada.

Embora essa afirmação seja válida quando se lança apenas um único dado, ela não se aplica quando o lançamento é composto pela soma de dois dados. Existem 36 possíveis combinações de resultados quando se lançam dois dados. Como resultado, a probabilidade de obter determinados valores varia.

Ao analisar a distribuição dos resultados, observamos que há uma probabilidade significativamente maior de obter valores de 6 e 7, enquanto a probabilidade de obter valores de 2 e 12 é menor. Portanto, a distribuição dos resultados não é uniforme quando a simulação envolve a soma de dois dados. Ela segue a distribuição normal como pode ser observado no histograma.
